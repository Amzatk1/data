```python
# Import libraries for model evaluation, visualization, feature scaling
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import ReduceLROnPlateau
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Dataset filepath 
data_path = '/Users/ayooluwakarim/Downloads/data'
train_file = f'{data_path}/training_bc_dataset.csv'
test_file = f'{data_path}/testing_bc_dataset.csv'

# To verify if files exist
if not os.path.exists(train_file) or not os.path.exists(test_file):
    print("File not found. Please check the file path and file names.")
else:
    train_data = pd.read_csv(train_file)
    test_data = pd.read_csv(test_file)
    print("Training Data Preview:")
    display(train_data.head())
    print("\nTesting Data Preview:")
    display(test_data.head())

    # Encode Diagnosis to binary where 0=Benign and 1=Malignant
    # Reference: Encoding labels conforms with data preprocessing guidelines in CRISP-DM. 
    # (Source: https://surreylearn.surrey.ac.uk/d2l/le/lessons/267715/topics/3165137.)
    train_data['Diagnosis'] = train_data['Diagnosis'].map({'Benign': 0, 'Malignant': 1})
    test_data['Diagnosis'] = test_data['Diagnosis'].map({'Benign': 0, 'Malignant': 1})

    # Define features (X) and target (y)
    X_train = train_data.drop(columns=['Diagnosis'])
    y_train = train_data['Diagnosis']
    X_test = test_data.drop(columns=['Diagnosis'])
    y_test = test_data['Diagnosis']

    # Feature Scaling with the aid of StandardScaler for better performance
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Reference: StandardScaler ensures features have zero mean and unit variance, 
    # therefore helping models converge faster and improve performance.
    # (Source: Scikit-learn documentation: https://scikit-learn.org/stable/modules/preprocessing.html#standardscaler)

    # SMOTE used for class imbalance
    smote = SMOTE()
    X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

    # Feature Selection using Recursive Feature Elimination
    rfe = RFE(estimator=LogisticRegression(max_iter=1000), n_features_to_select=10)
    X_train_rfe = rfe.fit_transform(X_train_balanced, y_train_balanced)
    X_test_rfe = rfe.transform(X_test_scaled)

    # Logistic Regression with GridSearchCV
    param_grid = {
        'C': [0.01, 0.1, 1, 10, 100],  # Regularization strengths utilized
        'solver': ['liblinear', 'lbfgs', 'saga']  # Solvers for optimization utilized
    }
    # To perform grid search to locate best parameters for use
    grid = GridSearchCV(LogisticRegression(max_iter=1000, class_weight='balanced'), param_grid, cv=5, scoring='roc_auc', verbose=1)
    grid.fit(X_train_rfe, y_train_balanced)

    # Evaluate Logistic Regression
    logistic_model = grid.best_estimator_

    # Logistic Regression Evaluation where prediction of classes of test data and prediction of probabilities for ROC
    y_pred_lr = logistic_model.predict(X_test_rfe)
    y_pred_prob_lr = logistic_model.predict_proba(X_test_rfe)[:, 1]

    # To calculate and display AUC and Confusion matrix figures
    roc_auc_lr = roc_auc_score(y_test, y_pred_prob_lr)
    conf_matrix_lr = confusion_matrix(y_test, y_pred_lr)

    print("\nLogistic Regression Performance:")
    print(f"AUC: {roc_auc_lr:.2f}")
    print("Confusion Matrix:\n", conf_matrix_lr)
    print("Classification Report:\n", classification_report(y_test, y_pred_lr))

    # Logistic Regression Confusion Matrix Plot
    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_matrix_lr, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
    plt.title("Confusion Matrix for Logistic Regression")
    plt.xlabel("Predicted")
    plt.ylabel("True Labels")
    plt.show()

    # Logistic Regression ROC Curve
    fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_prob_lr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr_lr, tpr_lr, label=f"Logistic Regression (AUC = {roc_auc_lr:.2f})")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve for Logistic Regression")
    plt.legend()
    plt.grid()
    plt.show()

    # To build and compile the optimized ANN model
    ann_model = Sequential([
        Dense(256, input_dim=X_train_rfe.shape[1], activation='relu', kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        Dense(128, activation='relu', kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        Dense(64, activation='relu', kernel_regularizer=l2(0.01)),
        Dense(1, activation='sigmoid')  # Sigmoid for binary classification
    ])
    # Compile ANN model with Adam optimizer, and lower learning rate to achieve stability
    ann_model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
    # Learning rate scheduler to adjust learning rate during training process
    lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)

    # Reference: ANN architectures with regularization techniques such as dropout are suitable 
    # for avoiding overfitting. (Source: Deep Learning by Ian Goodfellow et al., 2016)

    # Train the ANN model with training data
    history = ann_model.fit(
        X_train_rfe, y_train_balanced,
        epochs=200,
        batch_size=16,
        validation_split=0.2,
        verbose=1,
        callbacks=[lr_scheduler]
    )

    # Predictions and Threshold Tuning
    y_pred_prob_ann = ann_model.predict(X_test_rfe).ravel()
    threshold = 0.4  # Tuned threshold for better balance
    y_pred_ann = (y_pred_prob_ann > threshold).astype(int)

    # Evaluate ANN Model
    roc_auc_ann = roc_auc_score(y_test, y_pred_prob_ann)
    conf_matrix_ann = confusion_matrix(y_test, y_pred_ann)

    print("\nOptimized ANN Performance:")
    print(f"AUC: {roc_auc_ann:.2f}")
    print("Confusion Matrix:\n", conf_matrix_ann)
    print("Classification Report:\n", classification_report(y_test, y_pred_ann))

    # ANN Confusion Matrix Plot
    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_matrix_ann, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
    plt.title("Confusion Matrix for ANN")
    plt.xlabel("Predicted")
    plt.ylabel("True Labels")
    plt.show()
```
